````python
from matplotlib.patches import Circle, Wedge, Polygon
````

````python
fig, ax = plt.subplots(figsize = (4, 4))
p = Circle((0.5, 0.5), .2)
ax.add_patch(p)
````

````python
def circle_maker(x, y, r):
    fig, ax = plt.subplots(figsize = (4, 4))
    p = Circle((x, y), r, color = '#42f56f')
    ax.add_patch(p)
````

````python
def circle_maker2(n, r):
    fig, ax = plt.subplots(figsize = (4, 4))
    for i in range(n):
        x = np.random.random()
        y = np.random.random()
        p = Circle((x, y), r, color = '#42f56f')
        ax.add_patch(p)
````

````python
from ipywidgets import interact
import ipywidgets as widgets
````

````python
interact(circle_maker, x = widgets.FloatSlider(min = 0, max = 1, step = 0.1),
        y = widgets.FloatSlider(min = 0, max = 1, step = 0.1),
        r = widgets.FloatSlider(min = 0, max = 1, step = 0.1))
````

````python
interact(circle_maker2, n = widgets.IntSlider(min = 0, max = 100),
        r = widgets.FloatSlider(min = 0, max = 1, step = 0.01))
````

````python
ex = sns.load_dataset('flights')
````

````python
ex
````

````python
ex.year.value_counts()
````

````python
ex.info()
````

````python
ex.passengers.plot()
plt.axvline(40)
plt.axvline(100)
````

````python
def year_slider(start, stop):
    toplot = ex.copy()
    toplot = toplot.loc[ex['year'] >= start]
    toplot = toplot.loc[ex['year'] <= stop]
    plt.plot(toplot.passengers)
    return plt.show()
````

````python
interact(year_slider, start = widgets.IntSlider(min = ex.year.min(), max = ex.year.max()),
        stop = widgets.IntSlider(min = ex.year.min(), max = ex.year.max()))
````
