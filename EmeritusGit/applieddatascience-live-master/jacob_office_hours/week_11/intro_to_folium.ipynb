```python
noise = {'Noise - Street/Sidewalk', 'Noise - Residential', 'Noise - Vehicle'}
fire = {'Illegal Fireworks'}
cars = {'Illegal Parking', 'Blocked Driveway', 'Abandoned Vehicle'}
covid = {'NonCompliance with Phased Reopening'}

def color_mapper(x):
    if x in noise:
        return 'red'
    elif x in fire:
        return 'yellow'
    elif x in cars:
        return 'blue'
    elif x in covid:
        return 'purple'
    else:
        return 'green'
    
df['colors'] = df.complaint_type.apply(color_mapper)
```

```python
m = folium.Map(location=start_location, tiles = "StamenWatercolor")
def test_apply(x):
    popup = f'<strong>Department:</strong>{x[0]}\n\n<strong>Complaint Type:</strong>{x[1]}'
    folium.CircleMarker(location=[x[3], x[4]], popup = popup, color = x[-1], fill = True).add_to(m)
```

```python
df.dropna(subset = ['latitude', 'longitude']).apply(test_apply, axis = 1)
```
