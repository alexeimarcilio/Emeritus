 ### Introduction to Customizing Graphs
The customization of graphs in Matplotlib entails adding in aesthetic features to make your graphs and charts more appealing to your audience. Matplotlib has a multitude of ways to customize your graphs and charts and we will cover some of the more popular ones.

### Plot customizations
In this section we will talk about some plot customizations, including:
- adding a legend
- setting colors
- setting marker and line styles

Let's go through an example where we use several of the customizations identifed above

```python
# import pyplot
import matplotlib.pyplot as plt
# define our sample data for two data sets
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
# create a figure to render on
fig = plt.figure()
# creat our axes
ax = fig.add_axes([0,0,1,1])
# plot the two lines
l1 = ax.plot(x1,y,'rs-') # solid line with yellow colour and square marker
l2 = ax.plot(x2,y,'bo--') # dash line with green colour and circle marker
# add a legend
ax.legend(labels = ('IPhone', 'Samsung'), loc = 'lower right') # legend placed at lower right
# add a title
ax.set_title("Advertisement effect on Sales")
# create lables
ax.set_xlabel('# Ads')
ax.set_ylabel('sales in M$')
# show the plot
plt.show()
```

### Setting Limits
Matplotlib automatically arrives at the minimum and maximum values of variables to be displayed along x, y (and z axis in case of 3D plot) axes of a plot. However, it is possible to set the limits explicitly by using `set_xlim()` and `set_ylim()` functions.


```python
#import numpy
import numpy as np
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
# use arrange to generate our data
x = np.arange(1,10)
#plot x and exp(x) as y, color green
a1.plot(x, np.exp(x), 'g')
# don't set any limits
a1.set_title('auto limits')
plt.show()
```

Same plot now, but with the limits set:

```python
#import numpy
import numpy as np
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
# use arrange to generate our data
x = np.arange(1,10)
#plot x and exp(x) as y, color red
a1.plot(x, np.exp(x), 'r')
a1.set_title('explicitly set limits')
# set the limits to 10,000 and 10
a1.set_ylim(0,10000)
a1.set_xlim(0,10)
plt.show()
```

### Grids
The `grid()` function of axes object sets visibility of grid inside the figure to on or off. You can also display major / minor (or both) ticks of the grid. Additionally color, linestyle and linewidth properties can be set in the `grid()` function.

```python
# create the figure and subplots - there will be three of them
fig, axes = plt.subplots(1,3, figsize = (12,4))
# create a data set
x = np.arange(1,11)
# plot a line in subplot 1 with the grid on
axes[0].plot(x, x**3, 'g',lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')
# plot a line in subplot 2 with the custom grid on
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls = '-.', lw = 0.25)
axes[1].set_title('custom grid')
# plot a line in subplot 3 with NO grid
axes[2].plot(x,x)
axes[2].set_title('no grid')
# use side by side layout
fig.tight_layout()
plt.show()
```
