 ### Line Plots (Graphs)
   
A line plot is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments. Line graphs are used to track changes over short and long periods of time. Line graphs can also be used to compare data sets over the same period of time for more than one group (a series).
- X-axis requires quantitative variable
- Y-axis values must be numbers and can be set to be interpreted (formatted) as decimal, percent, and currency.
- Variables have contiguous values
- Familiar/conventional ordering among ordinals

## Data set for our line plot
For the data set for our line plots we will use the **numpy** library function ``arrange`` to give us generated data for plotting. Additionally, we will include **math**, which is the Python Math Library and provides us access to some common math functions and constants.

```python
import numpy as np
import math
# the x axis: ndarray object of angles between 0 and 2Ï€
x = np.arange(0, math.pi*2, 0.05)
# the y axis: sine values of angles in x to be displayed on y axis
y = np.sin(x)
```

## Single Series line plot
First, we create a simple line plot of our data using **pyplot**.

```python
import matplotlib.pyplot as plt

plt.plot(x,y)
plt.show()
```

Then we will add some labels and markers to make the plot presentable.

```python
#You can set the plot title, and labels for x and y axes.
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.plot(x,y)
plt.show()
```

## Multiple Series line plot
In this case we will add in the ``cosine`` as our second dataset so that we can demonstrate how to plot a line series. NOTE: we will use the same x ndarray, as we want to keep the same X-Axis values.
- use the same x values
- add the second y array (cosine)
- add the labels
- add a legend

```python
# the y axis for Cosine: cosine values of angles in x to be displayed on y axis
y_cos = np.cos(x)
#You can set the plot title, and labels for x and y axes.
plt.xlabel("angle")
plt.ylabel("sine/cosine values")
plt.title('sine/cosine wave')
plt.plot(x,y,x,y_cos)
plt.legend(['sin(x)', 'cos(x)'])      # legend entries as seperate strings in a list
plt.show()
```
