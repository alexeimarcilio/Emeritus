 ## What are 3D plots in Matplotlib
   
Three-dimensional plots are enabled by importing the **mplot3d** toolkit, included with the Matplotlib package. Three-dimensional axes can be created by passing the keyword `projection='3d'` to any of the normal axes creation routines. 

The 3D capabilitiy is provided by the **mplot3d toolkit** which simplifies 3D plotting capabilities in matplotlib by supplying an axes object that can create a 2D projection of a 3D scene. The resulting graph will have the same look and feel as regular 2D plots.


## Creating a scatter plot in 3D

The **mplot3d toolkit** allows us to make any 2D graph in **matplotlib** into a 3D graph.

The inclusion of ``mplot3d`` enables us to add the third dimension to the `sub_plot` and allows the plotly functions to "graph" on the 3rd axis.

```python
# libraries
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D
# include pyplot
import matplotlib.pyplot as plt
# include numpy
import numpy as np
# include pandas
import pandas as pd
 
# create the Dataset as a dataframe
df=pd.DataFrame({'X': range(1,201), 'Y': np.random.randn(200)*15+range(1,201), 'Z': (np.random.randn(200)*15+range(1,201))*2 })
 
# plot
fig = plt.figure(figsize=(8,8))
# use the 3D projection from the mplot3d import
ax = fig.add_subplot(111, projection='3d')
# not the scatterplot function will be a 3 axis graph
ax.scatter(df['X'], df['Y'], df['Z'], c='blue', s=60)
#show the plot
plt.show()
```

## Creating a bar chart in 3D
In this case we will:
- generate a set of data using ``meshgrid``
- calulate a Z or top value
- plot the 3d and 2d charts

```python
import numpy as np
import matplotlib.pyplot as plt
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import


# setup the figure and axes
fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(221, projection='3d')
ax2 = fig.add_subplot(222, projection='3d')
ax3 = fig.add_subplot(223) #2d
ax4 = fig.add_subplot(224) #2d

# fake data
_x = np.arange(4)
_y = np.arange(5)
# create a grid out of two arrays (cartesian indexing)
_xx, _yy = np.meshgrid(_x, _y)
# flatten the matrix into x and y arrays
x, y = _xx.ravel(), _yy.ravel()

#calculate a Z
top = x + y
bottom = np.zeros_like(top)
width = depth = 1

ax1.bar3d(x, y, bottom, width, depth, top, shade=True)
ax1.set_title('Shaded')

ax2.bar3d(x, y, bottom, width, depth, top, shade=False)
ax2.set_title('Not Shaded')

ax3.bar(x, top, color='red')
ax3.set_title('2D x and z')

ax4.bar(y, top, color='green')
ax4.set_title('2D y and z')


plt.show()
```
