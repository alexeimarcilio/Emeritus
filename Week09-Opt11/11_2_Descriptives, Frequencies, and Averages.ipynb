```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
sns.countplot(y=df['Age'])
```

```python
# For a smaller number of unique entries we can look at value counts for 
# each of the columns
# Can use this in the future
# Better than typing each variable
object_col = []
for column in df.columns:
    if df[column].dtype == object and len(df[column].unique()) <= 30:
        object_col.append(column)
        print(f"{column} : {df[column].unique()}")
        print(df[column].value_counts())
        print("*********************************")
        
```

```python
for column in df.columns:
    print(f"column: {column}: Number of unique values: {df[column].nunique()}")
```

```python
for column in df.columns:
    if df[column].dtype == object and len(df[column].unique()) <= 30:
        sns.countplot(df[column])
```

```python
# Find all the categorical features
object_col = []
for column in df.columns:
    if df[column].dtype == object and len(df[column].unique()) <= 5:
        object_col.append(column)


%matplotlib inline
        
fig, ax = plt.subplots(1, len(object_col), figsize=(20,5))
for i, object_col in enumerate(df[object_col]):
    df[object_col].value_counts().plot(kind="bar", ax=ax[i]).set_title(object_col)
#fig.show();

```
