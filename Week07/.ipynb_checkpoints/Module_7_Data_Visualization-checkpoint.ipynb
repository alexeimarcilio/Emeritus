## Box plots

A box plot is a plot that displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.

We will use the `boxplot()` function to create a boxplot with Matplotlib. Parameters in the `boxplot()` function determine orientation, colors, labels for the figure. 

[Back to top](#Index:) 

### Question 8
*10 points*

Using the data `data` below, create a box plot by using the function `plt.boxplot()`. To plot the outliers, set the parameter `flierprops` equal to `green_diamond`.

Make these customizations to your final plot:

- Title: "My box plot"
- X-label: "data"


Save your plot as a png with the name "plot8.png" in the folder "results".

````python
### GRADED

# Data
np.random.seed(123)
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))

# For outliers
green_diamond = dict(markerfacecolor='g', marker='D')

### YOUR SOLUTION HERE
fig = plt.figure()
# Create an axes instance
ax = fig.add_axes([0,0,1,1])
# Create the boxplot
plt.boxplot(data, flierprops=green_diamond)
# Display
#plt.show()
plt.title("My box plot")
plt.xlabel("data")
plt.savefig('results/plot8.png')
###
### YOUR CODE HERE
###
````

## Pair plots

A âpairs plotâ is also known as a scatter matrix. A scatter matrix is a collection of scatterplots organized into a grid (or matrix). Each scatterplot shows the relationship between a pair of variables. Plots are just elaborations on this, showing all variables paired with all the other variables. 

Here, you will create a pairs plot using pandas.

Create a pair plot of `iris_dataframe` using the pandas function `scatter_matrix` from the `plotting` module. Use these customizations inside the `pd.plotting.scatter_matrix` function:

````python
import numpy as np
# set the seed for the random function
np.random.seed(10)
# generate the sample sets: numpy.random.normal(loc=0.0, scale=1.0, size=None)
set_1 = np.random.normal(100, 10, 200)
set_2 = np.random.normal(80, 30, 200)
set_3 = np.random.normal(90, 20, 200)
set_4 = np.random.normal(70, 25, 200)
# create a collection of data to plot (list of ndarray)
data_to_plot = [set_1, set_2, set_3, set_4]
````

````python
# create a boxplot from the same data 
box=plt.boxplot(data_to_plot,vert=0,patch_artist=True,labels=['series1','series2','series3','series4'])
# select the colors for the boxes 
colors = ['cyan', 'blue', 'green', 'orange']
# assign the box colors
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
# display the box plot 
plt.show()
````
